```python
#Solve a System of Linear Equations
# Matrix A
A = [[2, 3],
     [1, 4]]

# Vector b
b = [8, 7]

# Extract values
a1 = A[0][0]
b1 = A[0][1]
a2 = A[1][0]
b2 = A[1][1]

c1 = b[0]
c2 = b[1]

# Determinant
D = a1*b2 - b1*a2

# Solve for x and y
x = (c1*b2 - b1*c2) / D
y = (a1*c2 - c1*a2) / D

print("Solution:")
print("x =", x)
print("y =", y)
```
